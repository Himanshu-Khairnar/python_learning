```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
arr1 = np.array([1,2,3,4,5])
arr2 = np.random.random((3,3))
arr3 = np.zeros((4,4))
arr4 = np.ones((4,4))


np.savez('arrays.npz', first=arr1, second=arr2, third=arr3,fourth=arr4)
loaded = np.load('arrays.npz')
print(loaded['first'])
print(loaded['second'])
print(loaded['third'])
print(loaded['fourth'])
```

```python
try:
    data=np.load('numpy-logo.npy')
    print(data)
    plt.figure(figsize=(6,6))
    plt.title("Numpy Logo")
    plt.imshow(1-data)
    plt.axis('off')
    plt.show()
except Exception as e:
    print("Error loading numpy-logo.npy:", e)
```
