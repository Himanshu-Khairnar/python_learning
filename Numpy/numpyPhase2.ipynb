### Numpy Array Operations

```python
import numpy as np
```

```python
arr = np.array([1,2,3,4,5,6,7,10])
print(f"Basic Slicing: {arr[1:4]} ")
print(f"Basic Slicing: {arr[1:8:2]} ")
print(f"Basic Slicing: {arr[-1]} ")
```

### 2d Array

```python
array_2d = np.array([[4, 2, 3], [1, 5, 6], [7, 8, 9]])
print(f"2d array: {array_2d[0][0:2]}")
```

```python
print(f"sort of array: {np.sort(array_2d,axis=1)}")
```

### Array compatibility

```python
a = np.array([1,2,3])
b = np.array([4,5,6])
c = np.array([7,8,9])

print(a.shape == b.shape)
```

```python
original = np.array([[1,2],[3,4]])
new_row = np.array([[5,6]])

with_new_row = np.vstack((original,new_row))
print(with_new_row)


new_col = np.array([[7],[8]])
with_new_col = np.hstack((original,new_col))
print(with_new_col)
```

```python
arr = np.array([1,2,3])
deleted = np.delete(arr,2)
print(deleted)
```
