```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
sales_data = np.array(
    [
        [1, 150000, 180000, 220000, 250000],
        [2, 120000, 140000, 160000, 190000],
        [3,200000,230000,260000,300000,],
        [4,180000,210000,240000,2700001,],
        [5, 160000, 185000, 205000, 230000],
    ],
)
print(sales_data.shape)
print(f"Sample data for first 3 res {sales_data[:3]} ")
```

### TOTAL and min sales per year

```python
print(np.sum(sales_data[:,1:], axis=0))

print(np.min(sales_data[:,1:],axis=1))
print(np.max(sales_data[:,1:],axis=0))
print(np.mean(sales_data[:,1:],axis=1))
print(np.cumsum(sales_data[:,1:],axis=1))
```

```python
cumsum = np.cumsum(sales_data[:,1:],axis=1)
plt.figure(figsize=(10,10))
plt.plot(np.mean(cumsum,axis=0))
plt.title("avg sales across all restaurant")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.ylim(0,1500000)
plt.xlim(0,4)
plt.grid(True)
plt.show()
```

```python
vector = np.array([[1,2,3]])
vector1 = np.array([6,7,8])
print(np.dot(vector,vector1))

vector = np.array([[1,2],[3,4]])
vector1 = np.array([[7,8],[10,11]])
print(np.dot(vector,vector1))

v1 = np.array([1,2,3])
v2 = np.array([4,5,6])



ang = math.acos(np.dot(v1,v2)/((np.linalg.norm(v1)*np.linalg.norm(v2))))
print(f"randians: {ang} angle: {math.degrees(ang)}")
```
